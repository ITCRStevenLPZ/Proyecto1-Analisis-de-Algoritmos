### Conceptos

#### Algoritmos Divide y Conquista

Este tipo de algotimos consisten en aplicar una técnica basada en tres puntos fundamentales. Las cuales son:<br>
<ol>
    <li>Dividir: Dividir el problema en sub-problemas mas sencillos de atacar.
    <li>Conquistar: Abordar recursivamente a los sub-problemas hasta estar resueltos.
    <li>Combinar: unir los subproblemas para así lograr obtener la respuesta requerida.
<ol>


#### Intercambio de  Adyacentes

Los algoritmos de ordenamiento del tipo intercambio de adyacentes, son aquellos que realizan un *swap* entre elementos que se encuentran adyacentes en un arreglo, siempre y cuando estos no satisfagan el ordenamiento deseado, de este manera después de ciertas iteraciones se obtendrá el arreglo ordenado.

### Análisis 

Dentro del análisis de los datos anteriores se abarcaran los comportamientos de ambos algoritmos para lograr identificar sus fortalezas y debilidades según sea el comportamiento de sus entradas, además estudiará el por que de estos resultados para poder explicar de una manera más teórica el por qué un algoritmo es más eficiente en algunos caso y en otros decae su rendimiento.

#### Análisis del Peor Caso

En el análisis de la sección del `Peor Caso` se puede observar que cuando los arreglos tienen un largo menor a 21 elementos el `insertion sort` es más eficiente, pero a partir de los 25 elementos el ganador indiscutible es el quicksort. Cabe rescatar que después de correr varias veces las simulación y experimentación, el punto de inflexión varía pero suele oscilar de entre 10 a 35 elementos, es decir el punto en el que el insertion sort pasa ser menos eficiente suele estar entre los 10 y los 35 elementos.

Al estar trabajando en una computadora en la cual no está única y exclusivamente corriendo la simulación, puede que los recursos de la computadora se desvíen a necesidades del sistema operativo y desarrollo de aplicaciones, sin embargo, el resultado aunque puede oscilar en el punto de inflexión es posible obtener una aproximación de en qué momento el `insertion sort` empieza a tender a ser menos eficiente temporalmente 

El algoritmo de ordenamiento Quicksort funciona realizando sub-listas hasta que el largo de estas sub-listas sea suficientemente corto como para que sea rápido y eficiente ordenarlas, esta particularidad del tipo de ordenamiento divide y conquista es una virtud y un defecto al mismo tiempo, por que al dividir las listas en sub-listas de un largo corto se obtienen más operaciones que cuando se hace un barrido con otro ordenamiento como insertion sort, es por esta razón que tanto en los siguientes experimentos como en este, vamos a ver que insertion sort va a ser mejor temporalmente con una cantidad `n` pequeña.

#### Análisis del Mejor Caso

Durante el estudio de las pruebas del mejor casos, se puede notar como el ganador indiscutible del enfrentamiento es el `insertion sort`, el estudio arroja un resultado inminente, en las pruebas de mejor caso el mejor algoritmo siempre será inserción en cualquier tamaño n que se escoja siendo n el largo del arreglo

Por la naturaleza de ambos algoritmos se puede denotar la razón del por qué este resultado, si bien es cierto en la teoría de puede vender el algoritmo Quicksort como uno de los mejores algoritmos de ordenamiento e insertion sort lo tenemos como un algoritmo medianamente eficiente pero con grandes desventajas en comparación al QuickSort, con esta premisa, ¿cómo es posible esta conclusión?, pues si estudiamos la naturaleza del algoritmo insertion sort y el caso de prueba es cuando este tenga el arreglo ordenado, como en este estudio, entonces podemos notar que siempre el insertion sort va a tener una complejidad de **n**, debido a que si está en orden lo que debe hacer es hacer una comparación por cada n y descubrir que no debe realizar ningún cambio. 

$a_1<a_2$ , $a_2<a_3$ , $a_3<a_4$ , $...$ , $a_{n-1} < a_n$

Además como se explicó en el experimento anterior el algoritmo de ordenamiento Quicksort por naturaleza debe dividir el arreglo en sub-arreglos y luego comparar los subarreglos pequeños para lograr denotar si está o no ordenado esto en el mejor caso de quicksort lo realiza en una complejidad temporal de $O(n \log n)$, como podemos comprobar es correcto que insertion sort es más eficiente cuando el mejor caso se presenta.
