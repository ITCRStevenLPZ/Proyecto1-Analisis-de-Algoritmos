# Solución a problema 3 #

## Problema 3

¿Cómo se compara el algoritmo quicksort con el algoritmo mergesort en términos de rendimiento temporal?

### Conceptos

#### QuickSort

Algoritmo de ordenamiento que pertenece a la sección de *Algoritmos de Divide y Conquista*, trabaja seleccionando un pivote y subdivide un arreglo de entrada en dos subarreglos, ordenando a la izquierda los elementos del arreglo menores al pivote y a la derecha los mayores y este proceso se vuelve a repetir recursivamente con cada uno de los subarreglos resultantes de la partición hasta tener un caso trivial.

```python
from Algoritmos import quickSort
Arreglo = [90,15,56,7,24,1]
quickSort(Arreglo)
print(Arreglo)
```

#### MergeSort

Algoritmo de ordenamiento que pertenece a la sección de *Algoritmos Divide y Venceras*, trabaja dividiendo un arreglo de entrada por la mitad una y otra vez hasta que cada subarreglo que genere quede de una longitud de 1 elemento, luego estos subarreglos se van concatenando en el orden dado.

```python
from Algoritmos import mergeSort
Arreglo2 = [123,1,56,89,15,8]
mergeSort(Arreglo2)
print (Arreglo2)
```

### Hipótesis

En este problema contaremos con 2 algoritmos en comparación. Primero **QuickSort**, con el cual contaremos con dos versiones de este: uno con pivote en el centro del array (pivote = $n/2$; siendo $n$ el tamaño del arreglo) y otro que se encuentre en el último elemento; identificados como *quickSortOP* y *quickSort* respectivamente en el archivo *Algoritmos.py*. 

Se espera que en el mejor caso de ambas versiones de QuickSort y del algoritmo **MergeSort** sean de complejidad temporal $O(nlogn)$ y que cumpla con las hipótesis de entradas que más adelante serán especificadas.

En el caso promedio se espera un rendimiento similar en ambos algoritmos QuickSort y MergeSort, aunque en este caso, QuickSort puede llegar a ser mas rápido, debido a que el factor de fusión es mayor, sin embargo tiene mucho que ver con la implementación de los algoritmos. Puede que si el arreglo este muy desordenado tienda a caer en el peor caso.

Para el peor caso se espera que MergeSort supere al QuickSort, el cual en teoría tiene una complejidad mucho mayor al peor caso que MergeSort, aunque varios estudios citan a QuickSort como ganador de esta comparación temporal (con MergeSort) debido a la naturaleza de su pivote, por lo que se espera que QuickSort sea más eficiente temporalmente dependiendo de un pivote específico.

**NOTA:** Debido a la naturaleza de esta sección, todo puede variar después de la experimentación y pueden haber contradicciones.


#### Hipótesis de pivotes

Teóricamente con QuickSort encontramos que:
* Es necesario saber que el **pivote** influye mucho en la complejidad de este algoritmo, especificamente en la posición del mismo en el arreglo:
    * Si el pivote se encuentra en el **primer** o **último** elemento o índice del arreglo de entrada, este va a tener la peor eficiencia posible en QuickSort.
    * Si el pivote se encuentra en un rango del **primer elemento + 1** hasta el **ultimo elemento -1** (siendo pivote de *media*, pivote *random* y pivote *mitad*) del arreglo de entrada, este va tener una mejor eficiencia.
    * Si estríctamente el pivote del arreglo se encuentra en el centro del arreglo **($n / 2$, siendo $n$ el tamaño total del arreglo)** este presentará la mejor eficiencia.

Teóricamente con MergeSort encontramos que:
* **NO** hay existencia de pivotes de una manera explícita. MergeSort se centra en la división recursiva de arreglos o sub-arreglos en la posición $n/2$ (siendo $n$ la cantidad total de elementos de un arreglo) hasta tener sub-arreglos triviales o fáciles de resolver, por eso se puede concluir:
    * La complejidad temporal va a depender precísamente de la estructura de las entradas, por lo que es factible ignorar tipos o variantes de MergeSort. Esto no quiere decir que no hallan variantes del mismo, sino más bien que no se contemplarán en esta comparación, debido a que no tienen tanta importancia como las variantes que hay con QuickSort.

#### Hipótesis de entradas y complejidades temporales

##### QuickSort

En lo que a entradas se refiere:

* El comportamiento del mejor caso del algoritmo de QuickSort ocurre cuando en cada paso de la recesión en la partición produce dos partes de igual longitud.*(a)*
* En el caso promedio: Para poder ordenar $n$ elementos en este caso el tiempo de ejecución es $O(nlog(n))$.Esto es porque la profundidad de recursión es $log(n)$ y en cada nivel hay $n$ elementos para ser procesados.*(b)*
* El peor caso ocurre cuando en cada paso de la recursión se produce una partición desbalanceada, principalmente esa parte consiste en un solo elemento y la otra parte consiste en el resto de los elementos. La profundidad de recursión es de $n-1$ y el tiempo de ejecución de quicksort es $O(n²)$.*(c)*


A continuación se presenta una imagen en la que se pueden ver gráficamente los casos de QuickSort:

![Casos de QuickSort](https://www.inf.hs-flensburg.de/lang/algorithmen/sortieren/quick/quickcpx.gif) 

##### MergeSort

En lo que a entradas se refiere:
* El mejor caso ocurre a la hora de que se divide el arreglo en dos sub-arreglos (división inicial) y se cumpla que el mayor valor del sub-arreglo de la izquierda sea **menor** al menor valor del sub-arreglo de la derecha. Esta característica no solo implica que el arreglo este ordenado exclusívamente, pueden haber arreglos que cumplan con esta característica y que no esten ordenados. Su complejidad temporal es $O(nlogn)$.
* El caso promedio ocurre cuando la lista está parcialmente ordenada o desordenada, con valores menores en sub-arreglos a la derecha y valores mayores en sub-arreglos a la izquierda, pero que los valores más grandes no se encuentren en sub-arreglos opuestos a la hora de hacer la fusión o merge. Su complejidad temporal es $O(nlogn)$.
* El peor caso de MergeSort ocurre cuando los dos valores más grandes del arreglo de entrada, en un paso de fusión, esten contenidos en sub-arreglos opuestos, causando que MergeSort tenga que continuar comparando elementos de lista en cada una de las listas opuestas hasta que se comparen los dos valores más grandes. Su complejidad temporal es $O(nlogn)$.

**NOTA:** En teoría MergeSort independiéntemente del arreglo en el que sea sometido o ejecutado siempre va a tener una complejidad temporal $O(nlogn)$, por lo que **en este problema** no es necesario hacer énfasis en los casos de este algoritmo.

### Análisis de Resultados

QuickSort es un algoritmo flexible, ya que este tiene variantes que lo pueden llegar a mejorar como el pivote. Queda demostrado que el pivote es clave para establecer el peor caso, ya que para que ocurra este, el pivote tiene que estar en el primer o en el último elemento del arreglo y el arreglo tiene que estar ordenado ascendentemente o descendentemente. La razón de este comportamiento es que QuickSort con pivote en los extremos de los arreglos tiene que recorrer la lista $(n-1)$ por cada llamada recursiva, generando al final una complejidad temporal $O(n^2)$, la cual es evidente en el Experimento 2. Por lo tanto para el desarrollo de los otros casos de QuickSort se deben establecer los pivotes correctos y las entradas necesarias para que las particiones generadas en el arreglo (los sub-arreglos) cumplan con la siguiente distribución:
![Casos de QuickSort](https://www.inf.hs-flensburg.de/lang/algorithmen/sortieren/quick/quickcpx.gif) 

En donde (a) es la distribución de las particiones ideal, que por consecuencia genera el mejor caso , (b) es la distribución en la cual se especifica el caso promedio y (c) es la distribución que genera la complejidad temporal (O(n^2)).

Para responder a la incógnita de este problema 3, MergeSort es más ineficiente que QuickSort, cuando este último tiene el pivote en la mitad debido a que MergeSort realiza mas operaciones de concatenación y comparación una vez que se divide el arreglo en sub-arreglos de tamaño 1, todo esto implica un aumento en la complejidad temporal, sin superar la complejidad $O(nlogn)$ como es evidente en el Experimento 4 y 5.

La siguiente imagen muestra de forma gráfica el funcionamiento de MergeSort:
![MergeSort](https://upload.wikimedia.org/wikipedia/commons/thumb/c/cc/Merge-sort-example-300px.gif/220px-Merge-sort-example-300px.gif) 

Pero si se compara MergeSort con la peor variante de QuickSort (pivote en el primer o en el último elemento del arreglo) el resultado es el contrario, por lo tanto para realizar una comparación más puntual es importante identificar o especificar con cual variante de QuickSort, MergeSort va a ser comparado.

Se cumplió con lo establecido en la hipótesis, ya se vio que para establecer los casos en QuickSort se debe tomar en cuenta a los pivotes y MergeSort independientemente de los valores de entrada va a tener una complejidad temporal igual debido a su funcionamiento, por consiguiente la siguiente tabla queda demostrada que es verdad y QuickSort es el ganador de esta comparación debido a su flexibilidad y mejor rendimiento con respecto a MergeSort.
