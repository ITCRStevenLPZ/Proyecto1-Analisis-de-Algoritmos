#### Análisis Bubble Sort

![BubbleSort](https://upload.wikimedia.org/wikipedia/commons/5/54/Sorting_bubblesort_anim.gif "Visualización BubbleSort")

**Comportamiento temporal con los resultados predichos teóricamente**

Para el ordenamiento burbuja encontramos que la complejidad obtenida fue de $O(n^2)$, para justificar este comportamiento podemos clarificar el comportamiento que tiene el `bubbleSort` sobre los casos presentes.

La complejidad cuadrática la podemos tomar a partir de diversos factores entre los cuales se encuentran:


1. Ciclo anidados: El algoritmo presenta dos ciclos `while` y `for` que realizan iteraciones para llevar cada número a su respectiva posición a través de la lista, una característica muy frecuente en complejidades cuadráticas. 


2. Condicion de finalización: Este ejemplo de algoritmo repite las iteraciones aun teniendo en cuenta que la lista se encuentre ordenada, por lo que tendremos la misma cantidad de comparaciones de posiciones `j` y `j+1`.


*Otras obervaciones:* Si analizamos los puntos anteriores y los datos obtenidos podemos notar un leve comportamiento de los tiempos entre "casos" pero bien sabemos que no se da una diferencia significativa. Por ejemplo podemos captar que los tiempos vistos en la tabla presentan en su **mayoría** la siguiente desigualdad:


$T_{peor}(n)<T_{promedio}(n)<T_{mejor}(n)$

Sin embargo a como vemos que el tamaño de $n$ aumenta, la diferencia entre tiempos se vuelve más grande. Podemos asumir lo mencionado por la entrada a la condicion del `if` y con ello realizar el `swap(j,j+1)`.
Nótese que las pruebas a veces imprimen un $T_{peor}(n)>T_{promedio}$ pero hay que recordar que se ejecutan en un ambiente donde hay demás procesos por lo que también eso influye en la consistencia de la desigualdad. 

En términos de complejidad esto no afecta por que para los tipos de listas siempre será cuadrática que los ciclos anidados siempre se cumplen. Un detalle importante es que la permutación de que la lista esté ordenada es casi improbable $\frac{1}{n!}$ y demasiado surrealista para tomar en cuenta y decir que bubblesort tiene tiempos cortos de ordenamiento, con ello podemos concluir que el algoritmo no presenta casos del todo y se espera un tiempo de $O(n^2)$.

#### Análisis de Quick Sort

![QuickSort](https://upload.wikimedia.org/wikipedia/commons/6/6a/Sorting_quicksort_anim.gif "Visualización QuickSort")

**Comportamiento temporal con los resultados predichos teóricamente**

Los datos obtenidos con el Quick Sort fueron semejantes a los propuestos en la hipótesis de la pregunta. Pudimos observar que dado un arreglo ordenado en cierta forma presenta un comportamiento diferente. Para el quicksort tuvimos que tomar en cuenta dos algoritmos diferentes para con consecuencia observar el impacto que tienen los pivotes en el rendimiento temporal.


Este algoritmo tiene dos puntos, el primero a favor para pertenecer a una notación linearítmica presentes en algoritmos *divide y conquista*. El segundo recae a una justificacion de lo que sucede en el peor caso. 


1. Divide la lista en sublistas, recursivamente hasta obtener listas nulas o de un elemento, dadas subdivisiones se dan por medio de un pivote. Los pivotes son los que determinan la cantidad de particiones que se realizan en el algoritmo en sí. Según observamos escoger el pivote a la mitad tiende a dividir a una partición de  $\mathcal \log_3(n)$ por que gráfica adjunta es más parecida con `linearitmica(n) = log(n,3)*n`


`pivote = (int)((L[i] + L[j]) / 2)` **QuicksortOP**

`pivot = arr[(int)(high)]`  **Quicksort**

Debemos notar que no sucede de la misma forma con el pivote al final, los datos indican la diferencia de elegir el pivote al final puede generar que para el peor caso se crea una complejidad cuadrática ya que se escoge un número que siempre es el mayor de las sublistas.

Podremos corroborar esto a consecuencia que `quick_peor` genera una lista ascendente por lo que los elementos de a_0 hasta a_{n-1} cumple siempren ` if   arr[j] <= pivot:` donde $j <n$ siempre. Teniendo por ende complejidad $O(n^2)$ con arreglos ordenados.


2. Realizar las respectivas llamadas de las sublistas a la función auxiliar `Quicksort` la cual según sea el caso coloca los números que sean menores o mayores en las diferentes sublistas al pivote. Consideremos que aunque hayan iteraciones anidadas,la cantidad de iteraciones máxima de $n$ es decir a la primera llamada $[a_0,a_1,a_2,a_3..a_{n-1}]$ por que  `j =n-1` e `i = n-1`.

Mencionado esto, las llamadas recursivas al final de la función son lo que causa la división de arreglos en tiempo logarítmico,estos a su vez en la mayoría de ocurrencias son más pequeños. Tomar en cuenta que la división tiende a hacer más grande cuando `first` y `j` tienen mayor diferencia, al igual que `i` con`last`. Esta diferencia es consecuencia de la calidad del pivote como determinante de cuanto se incrementa `i` o decrementa `j` en las iteraciones.


**Conclusiones**
El quicksort presenta una diferencia de comportamientos en base al pivote a elegir, las particiones que requieran hacer dependen ciertamente en si el pivote se encuentra de la media al resto de los elementos. Para los peores casos la complejidad si cumple con características cuadráticas por la cantidad de llamadas que realiza, inclusive para arreglos pequeños puede tener tiempos extensos.

Pensar en escenarios como que se elija el menor de los elementos en cada partición, vuelve a reiterar un caso muy poco probable para el algoritmo y tomarlo en cuenta puede ser muy falso. Podremos considerar al quick sort como uno de los mejores algoritmos de ordenamiento que presenta un comportamiento muy acercado al análisis teórico presentado en la hipótesis y también visto en clases.


#### Análisis de Merge Sort

![MergeSort](https://www.codeproject.com/KB/recipes/1087568/Merge_Sort__Double_Storage_.gif "Visualización MergeSort")

**Comportamiento temporal con los resultados predichos teóricamente**

El ejemplo del merge sort al ser nuevamente un algoritmo divide y conquista, partió con el único caso promedio de ser linearítmico, algo esperado en la hipótesis. Se obtuvieron resultados de orden linearítmico, para definir este tiempo debemos considerar los factores siguientes:

1. Divide la lista en sublistas, una función auxiliar que divide en mitades la lista recursivamente hasta obtener listas nulas o de un elemento, dado como resultado el factor logaritmico de $T(n)$. Esto es muy usual en algoritmos de divide en conquista, usualmente presentan llamadas que parten a problemas más pequeños.


`m = (l+(r-1))//2` 

`mergesort(arr, l, m)`
  
`mergesort(arr, m+1, r)`
  
`merge(arr, l, m, r)`

Notemos que estas líneas se ejecutan hasta que los índices no den para dividir más sublistas, la complejidad logarítmica la encontramos aquí, dado que se divide en problemas pequeños a una reducción $\mathcal \log_2(n)$

2. Realizar las respectivas llamadas de las sublistas a la función de `merge` la cual encola y desencola las mitades segun sea el caso que los números sean menores o mayores. Debemos tomar que `merge` no presenta anidación de casos y todos los ciclos se cumplen independiente entre ellos.


`for i in range(0 , n1):`

`for j in range(0 , n2):`

`while i < n1 and j < n2 :`

Podemos observar que las copias de los arreglos `n1 y n2`, seran ordenados de tal forma que queden menores y mayores respectivamente. Conforme avance el algoritmo y las llamadas notaremos que las iteraciones totales (n1+n2) serán de cantidad $|n-k|$ donde $k$ es una diferencia del tamaño entre los arreglos.

`while i < n1:`
`while j < n2:`
